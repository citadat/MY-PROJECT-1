```python
def get_employee_details(employee_name, employees):
    formatted_name = employee_name.title()

    try:
        employee_details = employees[formatted_name]
        return employee_details
    except KeyError:
        return None  # Return None if employee not found

def process_salary_data(salary_data):
    try:
        # Print the original salary data
        print("Original Salary Data:")
        for employee, salary in salary_data.items():
            print(f"{employee}: ${salary}")

        # Calculate and print average salary
        total_salary = sum(salary_data.values())
        average_salary = total_salary / len(salary_data)
        print("\nAverage Salary:", "${:,.2f}".format(average_salary))

        # Find and print the highest paid employee
        highest_paid_employee = max(salary_data, key=salary_data.get)
        print("\nHighest Paid Employee:", highest_paid_employee)
        print("Salary:", "${:,.2f}".format(salary_data[highest_paid_employee]))

    except ZeroDivisionError:
        print("Error: Division by zero. Ensure there are employees in the salary data.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
employees_data = {
    'GARY JIMENEZ': {'id': 1, 'position': 'CAPTAIN III POLICE DEPARTMENT'},
    'ALBERT PARDINI': {'id': 2, 'position': 'CAPTAIN III POLICE DEPARTMENT'},
    'CHRISTOPHER CHONG': {'id': 3, 'position': 'WIRE ROPE CABLE MAINTENANCE MECHANIC'},
    # Add more employees as needed
}

salary_data = {
    'GARY JIMENEZ': 538909.28,
    'ALBERT PARDINI': 335279.91,
    'CHRISTOPHER CHONG': 332343.61,
    # Add more employees and salaries as needed
}

try:
    employee_name_input = input("Enter employee name: ")
    employee_details = get_employee_details(employee_name_input, employees_data)

    if employee_details:
        print("\nEmployee Details:")
        for key, value in employee_details.items():
            print(f"{key}: {value}")

    process_salary_data(salary_data)

except KeyboardInterrupt:
    print("\nProgram terminated by the user.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
