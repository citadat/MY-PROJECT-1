REMOVE DECIMAL FROM THE FLOAT

```python
df = pd.DataFrame(healthcare_fee)
```

```python
df['bmi'] = df['bmi'].astype(int)
```

```python
sex_mapping ={'female': 1, 'male': 2}
df['sex'] = df['sex'].replace(sex_mapping)

smoker_mapping ={'yes': 1, 'no': 2}
df['smoker'] = df['smoker'].replace(smoker_mapping)

region_mapping ={'southeast': 1, 'southwest': 2, 'northwest': 3, 'northeast': 4}
df['region'] = df['region'].replace(region_mapping)
df
```

FIND THE OUTLIERS. BUT ARE GOING TO DO NOTHING ABOUT IT BECAUSE THEY ARE HEALTH PARAMETERS.

```python
import numpy as np
```

```python
import numpy as np
import pandas as pd

# Assuming df is a pandas DataFrame
# Calculate quartiles and IQR for a specific column 'column_name'
Q1 = np.percentile(df['age'], 25)
Q3 = np.percentile(df['age'], 75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = np.where((df['age'] < lower_bound) | (df['age'] > upper_bound))[0]

print("Indices of outliers:", outliers)
print("Outlier values:", df['age'][outliers])
```
