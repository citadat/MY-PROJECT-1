```python
class PolicyProduct:
    def __init__(self, product_id, name, coverage_type, premium):
        self.product_id = product_id
        self.name = name
        self.coverage_type = coverage_type
        self.premium = premium
        self.status = "Active"

    def display_product_details(self):
        print(f"Product ID: {self.product_id}")
        print(f"Name: {self.name}")
        print(f"Coverage Type: {self.coverage_type}")
        print(f"Premium: ${self.premium}")
        print(f"Status: {self.status}")
        print("\n")

    def update_product(self, name=None, coverage_type=None, premium=None):
        if name is not None:
            self.name = name
        if coverage_type is not None:
            self.coverage_type = coverage_type
        if premium is not None:
            self.premium = premium
        print(f"Product {self.product_id} updated successfully.")

    def suspend_product(self):
        if self.status == "Active":
            print(f"Suspending product {self.product_id}")
            self.status = "Suspended"
        else:
            print(f"Product {self.product_id} is already suspended.")

    def reactivate_product(self):
        if self.status == "Suspended":
            print(f"Reactivating product {self.product_id}")
            self.status = "Active"
        else:
            print(f"Product {self.product_id} is already active.")


# Creating instances of PolicyProduct
product1 = PolicyProduct(1, "Life Insurance", "Life Coverage", 100)
product2 = PolicyProduct(2, "Auto Insurance", "Vehicle Coverage", 50)

# Displaying initial product details
product1.display_product_details()
product2.display_product_details()

# Updating and suspending products
product1.update_product(premium=120)
product1.display_product_details()

product2.suspend_product()
product2.display_product_details()
```
