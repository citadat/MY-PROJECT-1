```python
diagnosis_mapping ={'B': 1, 'M': 2}
BreastCancer['diagnosis'] = BreastCancer['diagnosis'].replace(diagnosis_mapping)
BreastCancer
```

```python
BreastCancer.corr()
```

```python
corl = BreastCancer.corr()
corl
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 9)) # Adjust the figure size as needed

sns.heatmap(corl, cmap="coolwarm", annot=True, fmt=".2f", annot_kws={"size": 6}, linewidths=.5)

# Add labels and adjust the font sizes
plt.title("Correlation Heatmap", fontsize=20)
plt.xlabel("Features", fontsize=10)
plt.ylabel("Features", fontsize=10)
```

```python
x=corl.drop(columns=["diagnosis"])
y=corl[["diagnosis"]]
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=Sequential()
model.add(Dense(15,activation='relu', input_dim=30))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=200,validation_split=0.2)
```

```python
tr_loss = history.history['loss']
val_loss = history.history['val_loss']
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

Epochs = [i+1 for i in range(len(tr_loss))]
loss_label = f'best epoch= {str(index_loss + 1)}'

# Plot training history
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```
