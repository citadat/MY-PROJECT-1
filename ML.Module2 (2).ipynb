```python
mean_value = Telemedicine_use["Value"].mean()
Telemedicine_use["Value"].fillna(mean_value, inplace=True)
```

```python
mean_low_ci = Telemedicine_use["Low CI"].mean()
Telemedicine_use["Low CI"].fillna(mean_low_ci, inplace=True)
```

```python
mean_high_ci = Telemedicine_use["High CI"].mean()
Telemedicine_use["High CI"].fillna(mean_high_ci, inplace=True)
```

```python
# We need to split the confidence interval into two separate columns and then convert them to float
Telemedicine_use[['Lower_CI', 'Upper_CI']] = Telemedicine_use["Confidence Interval"].str.replace('%', '').str.split(' - ', expand=True).astype(float)

# Then we can fill the NaN values with the mean of the confidence interval
mean_confidence_interval = (Telemedicine_use['Lower_CI'] + Telemedicine_use['Upper_CI']) / 2
Telemedicine_use["Confidence Interval"].fillna(mean_confidence_interval, inplace=True)
```

```python
# Convert the "Quartile Range" column to numeric values
Telemedicine_use["Quartile Range"] = pd.to_numeric(Telemedicine_use["Quartile Range"], errors='coerce')

# Calculate the mean and fill NaN values with the mean
mean_quartile_range = Telemedicine_use["Quartile Range"].mean()
Telemedicine_use["Quartile Range"].fillna(mean_quartile_range, inplace=True)
```

```python
def remove_outliers(df, cols):
    for col in cols:
        df = df[(df[col] - df[col].mean()) / df[col].std() < 3]
    return df
numeric_cols = ['Time Period', 'Value', 'Low CI', 'High CI']
Telemedicine_use = remove_outliers(Telemedicine_use, numeric_cols)
```
